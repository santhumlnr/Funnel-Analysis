```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample funnel data (replace with actual data)
data = {
    "user_id": [1, 2, 3, 4, 5, 6, 7, 8],
    "stage": ["Visit Homepage", "Visit Homepage", "Sign Up", "Sign Up", 
              "Add to Cart", "Add to Cart", "Checkout", "Complete Purchase"],
}

# Convert data to a DataFrame
df = pd.DataFrame(data)

# Count users at each stage
funnel_counts = df['stage'].value_counts()

# Sort stages logically
stages = ["Visit Homepage", "Sign Up", "Add to Cart", "Checkout", "Complete Purchase"]
funnel_counts = funnel_counts.reindex(stages)

# Visualization: Funnel chart
plt.figure(figsize=(8, 6))
funnel_counts.plot(kind='bar', color='orange')
plt.title("User Funnel Analysis")
plt.xlabel("Stages")
plt.ylabel("Number of Users")
plt.xticks(rotation=15)
plt.grid(axis='y')
plt.show()

# Print insights
print("Funnel Analysis Insights:")
print("-" * 30)
print(funnel_counts)
```
